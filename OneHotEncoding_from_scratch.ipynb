```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

#First of all  lets transform the categorical data to numeric using one hot encoding.

```python
class OneHotEncoder:
    def __init__(self):
        pass

    def fit(self, column_array):
        self._categories=np.unique(column_array)

    def transform(self,column_array):
        result=[]
        for value in column_array:
            row=[]
            for element in self._categories:
                if value==element:
                    row.append(1)
                else:
                    row.append(0)
            result.append(row)
        return np.array(result)
    
    def fit_transform(self, column_array):
        self.fit(column_array)
        return self.transform(column_array)

```

```python
onehot1=OneHotEncoder()
onehot2=OneHotEncoder()
onehot3=OneHotEncoder()
onehot4=OneHotEncoder()
```

```python
tips.columns
```

```python
encoded_sex=onehot1.fit_transform(tips['sex'])
encoded_smoker=onehot2.fit_transform(tips['smoker'])
encoded_day=onehot3.fit_transform(tips['day'])
encoded_time=onehot4.fit_transform(tips['time'])
```

#lets see the example of encoded day

```python
encoded_day
```

```python
print(onehot1._categories)
print(onehot2._categories)
print(onehot3._categories)
print(onehot4._categories)
```

#concatating the encoded values to original array

```python
encoded=np.hstack((tips.values, encoded_sex, encoded_smoker,encoded_day, encoded_time))
encoded
```

#removing the encoded categorical datas

```python
encoded=np.delete(encoded, np.s_[2:6], axis=1)
encoded
```
